```python
#EDA Exploratary

df1=data1
df2=data2
df1.index=df1.Timestamp
df2.index=df2.Timestamp
```

```python
%matplotlib inline
plt.figure(figsize=(20,6))
plt.subplots_adjust(hspace= .2)
plt.subplot(1,2,1)
plt.title("highest weighted price",fontsize=12)
plt.xlabel("Timestamp",fontsize=12)
plt.ylabel("Weighted_Price")
sns.lineplot(x = df1.index, y = df1.Weighted_Price,color="orange")

plt.subplots_adjust(hspace=0.3)
plt.subplot(1,2,2)
plt.title("highest weighted price",fontsize=12)
plt.xlabel("Timestamp",fontsize=12)
plt.ylabel("weighted_price")
sns.lineplot(x = df2.index, y = df2.Weighted_Price,color="red")

plt.figure(figsize=(20,6))
plt.title("comaprison",fontsize=12)
plt.xlabel("Timestamp",fontsize=12)
plt.ylabel("weighted_price")

sns.lineplot(x = df1.index, y = df1.Weighted_Price,color="orange")
sns.lineplot( x= df2.index, y = df2.Weighted_Price,color="red")

plt.show()
```

```python
%matplotlib inline
plt.figure(figsize=(20,6))
bar=sns.stripplot(x="Timestamp",y="High",data=data1,hue="Weighted_Price")
bar.set_xlabel("timestamp")
plt.set_ylabel("high")
plt.show()
```
